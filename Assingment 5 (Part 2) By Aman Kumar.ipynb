#### Data Visualization

```python
from ipywidgets import interact
```

```python
def interact_plot(variable):
    plt.figure(figsize=(10,6))
    plt.plot(df['Janurary'], df[variable])
    plt.title(f"{variable} Over Janurary")
    plt.xlabel("Date")
    plt.ylabel(f"{variable}")
    plt.grid(True)
    plt.show()
```

```python
interact(interact_plot, variable = ['Temperature_Max(°F)', 'Temperature_Avg', 'Temperature_Min', 'pm25',
       'pm10', 'o3', 'no2', 'so2', ' co', 'Humidity_Max(%)', 'Humidity_AVg(%)',
       'Humidity_Min(%)', 'Wind_Speed_Max(mph)', 'Wind_Speed_Avgmph)',
       'Wind_Speed_Min(mph)', 'Precipitation (in)'])
```

#### Correlation heatmap

```python
plt.figure(figsize=(12,12))
sns.heatmap(df.corr(), cmap='Paired')
```

```python
# Scatter plot for Precipitation vs. Avg. Humidity
sns.scatterplot(data=df, x='Precipitation (in)', y='Humidity_AVg(%)')
plt.title('Precipitation vs. Avg. Humidity')
```

```python
# Scatter plot for Air Quality vs. Temperature
sns.scatterplot(data=df, x='pm25', y='Temperature_Avg')
plt.title('pm25 vs. Temperature')
```

```python
# Scatter plot for Wind_Speed_Avg vs. Temperature
sns.scatterplot(data=df, x='Wind_Speed_Avgmph)', y='Temperature_Avg')
plt.title('Wind_Speed_Avg vs. Temperature')
```

```python
# Scatter plot for Humidity_AVg(%) vs. Temperature
sns.scatterplot(data=df, x='Humidity_AVg(%)', y='Temperature_Avg')
plt.title('Humidity_AVg(%) vs. Temperature')
```
